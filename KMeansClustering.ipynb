```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv("student_clustering.csv")
```

```python
df.shape
```

```python
df
```

```python
plt.scatter(df['cgpa'], df['iq'])
```

```python
wcss=[]
for i in range(1,11):
    km= KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
X=df.iloc[:,:].values
km= KMeans(n_clusters=4)
y_mean=km.fit_predict(X)
```

```python
#rows in cluster 0 ,
X[y_mean==0]
```

```python
plt.scatter(X[y_mean==0,0],X[y_mean==0,1], color="blue")
plt.scatter(X[y_mean==1,0],X[y_mean==1,1], color="red")
plt.scatter(X[y_mean==2,0],X[y_mean==2,1], color="green")
plt.scatter(X[y_mean==3,0],X[y_mean==3,1], color='yellow')
```

```python
from sklearn.datasets import make_blobs
```

```python
centroids=[(-5, -5,5),(5,5,-5),(3.5,-2.5,4),(-2.5 , 2.5, -4)]
cluster_std=[1,1,1,1]
x, y = make_blobs(n_samples=200 , cluster_std=cluster_std, centers=centroids, n_features=3 , random_state=1)
```

```python
import plotly.express as px
```

```python

# ✅ Now 3D scatter plot will work
fig = px.scatter_3d(
    x=X[:, 0],
    y=X[:, 1],
    z=X[:, 2],
    color=y.astype(str),  # color by cluster
    title="3D Scatter Plot of Clusters"
)
fig.show()
```

```python
knn= KMeans(n_clusters=4)
y_pred=knn.fit_predict(x)
```

```python
df=pd.DataFrame()
```

```python
df['col1']=x[:,0]
df['col2']=x[:, 1]
df['col3']=x[:,2]
df['label']=y_pred
```

```python
fig= px.scatter_3d(df, x='col1', y='col2', z='col3', color='label')
fig.show()
```
