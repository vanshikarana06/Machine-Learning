Pipelines chains together mutiple steps so that the output of each step is used as input to next step .
pipelines make it easy to apply the same preprocessing to tain and test

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
df= pd.read_csv(r"C:\Users\sus\Downloads\train.csv")
```

```python
#Aim : train a a model so that i could predict if a member will servive or not  
```

```python
df.head(2)
```

```python
df=df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
```

```python
X_train, X_test , y_train , y_test = train_test_split(df.drop(columns=["Survived"]), df["Survived"], test_size=0.2, random_state=42)
```

```python
X_train.head(2)
```

```python
X_test.head(2)
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
#Applying imputation for missing values
si_age= SimpleImputer()# deafault strategy is mean
si_embarked= SimpleImputer(strategy="most_frequent")

X_train_age= si_age.fit_transform(X_train[["Age"]])
X_train_embarked= si_embarked.fit_transform(X_train[["Embarked"]])

X_test_age= si_age.transform(X_test[["Age"]])
X_test_embarked= si_embarked.transform(X_test[["Embarked"]])
   
```

```python
# Create encoder objects

# one hot encoding Sex and Embarked

ohe_sex = OneHotEncoder(sparse_output = False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output = False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train_sex   # we didn'tremove the fisrt column to remove multicoliniarity as we are going to 
#implement decision tree not any linear model.
```

```python
# now we are going to concatenate the requied cols 
```

```python
X_train_rem= X_train.drop(columns=['Sex', 'Age', 'Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex', 'Age', 'Embarked'])
```

```python
X_train_transform = np.concatenate((X_train_rem , X_train_age, X_train_sex, X_train_embarked), axis=1)
X_test_transform = np.concatenate((X_test_rem , X_test_age, X_test_sex, X_test_embarked), axis=1)
```

```python
X_train_transform.shape
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train_transform , y_train)
```

```python
y_pred=clf.predict(X_test_transform)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
