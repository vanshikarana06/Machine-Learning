```python

df = pd.read_csv(r"C:\Users\sus\Downloads\play_tennis.csv")


df.drop(columns=['day'], inplace=True)

# Define the conditions to predict.
conditions = {
    'outlook': 'Sunny',
    'temp': 'Hot',
    'humidity': 'High',
    'wind': 'Weak'
}

# 1. Calculate prior probabilities P(Yes) and P(No)
total_rows = len(df)
yes_rows = len(df[df['play'] == 'Yes'])
no_rows = len(df[df['play'] == 'No'])

p_yes = yes_rows / total_rows
p_no = no_rows / total_rows

# 2. Calculate conditional probabilities for each feature given the 'play' outcome.
# P(feature=value | play=Yes)
p_outlook_sunny_given_yes = len(df[(df['outlook'] == 'Sunny') & (df['play'] == 'Yes')]) / yes_rows if yes_rows > 0 else 0
p_temp_hot_given_yes = len(df[(df['temp'] == 'Hot') & (df['play'] == 'Yes')]) / yes_rows if yes_rows > 0 else 0
p_humidity_high_given_yes = len(df[(df['humidity'] == 'High') & (df['play'] == 'Yes')]) / yes_rows if yes_rows > 0 else 0
p_wind_weak_given_yes = len(df[(df['wind'] == 'Weak') & (df['play'] == 'Yes')]) / yes_rows if yes_rows > 0 else 0

# P(feature=value | play=No)
p_outlook_sunny_given_no = len(df[(df['outlook'] == 'Sunny') & (df['play'] == 'No')]) / no_rows if no_rows > 0 else 0
p_temp_hot_given_no = len(df[(df['temp'] == 'Hot') & (df['play'] == 'No')]) / no_rows if no_rows > 0 else 0
p_humidity_high_given_no = len(df[(df['humidity'] == 'High') & (df['play'] == 'No')]) / no_rows if no_rows > 0 else 0
p_wind_weak_given_no = len(df[(df['wind'] == 'Weak') & (df['play'] == 'No')]) / no_rows if no_rows > 0 else 0

# 3. Calculate the numerator for the posterior probabilities (unnormalized).
numerator_yes = p_yes * p_outlook_sunny_given_yes * p_temp_hot_given_yes * p_humidity_high_given_yes * p_wind_weak_given_yes
numerator_no = p_no * p_outlook_sunny_given_no * p_temp_hot_given_no * p_humidity_high_given_no * p_wind_weak_given_no

# 4. Normalize the probabilities.
total_numerator = numerator_yes + numerator_no
prob_yes = numerator_yes / total_numerator
prob_no = numerator_no / total_numerator

# 5. Make the final prediction based on the highest probability.
prediction = 'Yes' if prob_yes > prob_no else 'No'

# Print the prediction and the probabilities.
print(f"For the conditions: outlook=Sunny, temp=Hot, humidity=High, wind=Weak")
print(f"The manual Naive Bayes model predicts 'play' will be: {prediction}")
print(f"Probabilities:")
print(f"  Yes: {prob_yes:.2f}")
print(f"  No: {prob_no:.2f}")

# Example of the calculations for comparison
print("\nIntermediate calculations for the 'No' outcome:")
print(f"P(No) = {p_no:.2f}")
print(f"P(outlook=Sunny|No) = {p_outlook_sunny_given_no:.2f}")
print(f"P(temp=Hot|No) = {p_temp_hot_given_no:.2f}")
print(f"P(humidity=High|No) = {p_humidity_high_given_no:.2f}")
print(f"P(wind=Weak|No) = {p_wind_weak_given_no:.2f}")
print(f"Numerator(No) = {numerator_no:.4f}")

print("\nIntermediate calculations for the 'Yes' outcome:")
print(f"P(Yes) = {p_yes:.2f}")
print(f"P(outlook=Sunny|Yes) = {p_outlook_sunny_given_yes:.2f}")
print(f"P(temp=Hot|Yes) = {p_temp_hot_given_yes:.2f}")
print(f"P(humidity=High|Yes) = {p_humidity_high_given_yes:.2f}")
print(f"P(wind=Weak|Yes) = {p_wind_weak_given_yes:.2f}")
print(f"Numerator(Yes) = {numerator_yes:.4f}")
```

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
import numpy as np

# The code reads the data from a CSV file named 'play_tennis.csv'.
# Make sure you have a 'play_tennis.csv' file in the same directory.
try:
    df = pd.read_csv(r'C:\Users\sus\Downloads\play_tennis.csv')
except FileNotFoundError:
    print("Error: The file 'play_tennis.csv' was not found.")
    print("Please create the file and populate it with the provided data.")
    exit()

# Drop the 'day' column as it's not needed for the prediction.
df.drop(columns=['day'], inplace=True)

# Create a dictionary to hold a separate LabelEncoder for each column.
encoders = {}
categorical_cols = ['outlook', 'temp', 'humidity', 'wind', 'play']

for col in categorical_cols:
    encoders[col] = LabelEncoder()
    df[col] = encoders[col].fit_transform(df[col])

# Define features (X) and target (y).
X = df.drop('play', axis=1)
y = df['play']

# Initialize and train the Naive Bayes model.
model = GaussianNB()
model.fit(X, y)

# Define the new data point to predict.
new_data = pd.DataFrame({
    'outlook': [encoders['outlook'].transform(['Sunny'])[0]],
    'temp': [encoders['temp'].transform(['Hot'])[0]],
    'humidity': [encoders['humidity'].transform(['High'])[0]],
    'wind': [encoders['wind'].transform(['Weak'])[0]]
})

# Make a prediction.
prediction_encoded = model.predict(new_data)
# Use the 'play' encoder to inverse transform the prediction.
prediction_label = encoders['play'].inverse_transform(prediction_encoded)

# Get the prediction probabilities.
prediction_proba = model.predict_proba(new_data)
# Get the class labels from the 'play' encoder.
class_labels = encoders['play'].classes_

# Print the prediction and the probabilities.
print(f"For the conditions: outlook=Sunny, temp=Hot, humidity=High, wind=Weak")
print(f"The scikit-learn Naive Bayes model predicts 'play' will be: {prediction_label[0]}")
print(f"Probabilities:")
for i, label in enumerate(class_labels):
    # Format the probability to two decimal places.
    print(f"  {label}: {prediction_proba[0][i]:.2f}")
```
