```python
import os
from PIL import Image
import matplotlib.pyplot as plt

folder_path = r"C:\Users\sus\Downloads\pokemon_jpg\pokemon_jpg"

def sort_key(filename):
    name = filename.split('.')[0]
    return int(name) if name.isdigit() else float('inf')  # Non-numeric filenames go last

image_list = []

# Load and sort filenames
for filename in sorted(os.listdir(folder_path), key=sort_key):
    if filename.lower().endswith(".jpg"):
        try:
            img_path = os.path.join(folder_path, filename)
            image = Image.open(img_path)
            image_list.append((filename, image))
            print(f"Loaded: {filename}")
        except Exception as e:
            print(f"Error loading {filename}: {e}")

# Display first 5 images using matplotlib
for i in range(min(5, len(image_list))):
    name, img = image_list[i]
    plt.figure(figsize=(2, 2))
    plt.imshow(img)
    plt.title(name)
    plt.axis("off")
    plt.show()


```
