# woking with date


```python
date["date"]= pd.to_datetime(date["date"])
```

```python
date.info()
```

```python
# extract year
date["date_year"]= date['date'].dt.year
```

```python
date
```

```python
date["date_day"]=date["date"].dt.day
```

```python
date.head(2)
```

```python
date["day_of_week"]= date["date"].dt.day_name()
```

```python
date.head(1)
```

```python
import numpy as np
date["date_is_weekend"] = np.where(date["day_of_week"].isin(["Sunday", "Saturday"]), 1, 0)
```

```python
date
```

```python
date["date_week"] = date["date"].dt.isocalendar().week
```

```python
date
```

```python
date["quarter"]=date["date"].dt.quarter
date["semester"]=np.where(date["quarter"].isin([1,2]),1,2)
```

```python
date
```

```python
# extract time elapsed between dates 
import datetime
today= datetime.datetime.today()
today
```

```python
today-date["date"]
```

```python
(today-date["date"]).dt.days
```

```python
# months passed
months_passed = (today.year - date["date"].dt.year) * 12 + (today.month - date["date"].dt.month)
date["months_passed"] = months_passed
```
