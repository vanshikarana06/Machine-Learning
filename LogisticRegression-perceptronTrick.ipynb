# Perceptron learning 

```python
from sklearn.datasets import make_classification

x, y = make_classification(
    n_samples=100,
    n_features=2,       # only 2 features → can plot in 2D
    n_informative=1,
    n_redundant=0,
    n_classes=2,
    n_clusters_per_class=1,
    class_sep=10,       # makes points well-separated
    random_state=42,
    hypercube=False
)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1], c=y, cmap="winter", s=100)
plt.show()
```

```python

def step(z):
    return 1 if z > 0 else 0

def perceptron(X, y):
    X = np.insert(X, 0, 1, axis=1)   # add bias column
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0, 100)
        y_hat = step(np.dot(X[j], weights))
        weights = weights + lr * (y[j] - y_hat) * X[j]
    
    b = weights[0]
    w1 = weights[1]
    w2 = weights[2]
    
    return b, w1, w2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Train perceptron
b, w1, w2 = perceptron(x, y)

plt.figure(figsize=(10,6))

# Plot the two classes with different colors
plt.scatter(x[y==0, 0], x[y==0, 1], color="blue", label="Class 0", s=100)
plt.scatter(x[y==1, 0], x[y==1, 1], color="green", label="Class 1", s=100)

# Plot decision boundary
x_vals = np.linspace(x[:,0].min(), x[:,0].max(), 100)
y_vals = -(w1 * x_vals + b) / w2
plt.plot(x_vals, y_vals, color='red', linewidth=3, label="Decision Boundary")

plt.ylim(-3, 2)
plt.legend()
plt.show()



```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def perceptron_sigmoid(X, y, lr=0.1, epochs=200):
    """
    Perceptron with sigmoid activation (like single-layer logistic regression)
    
    Parameters:
    X : np.array, shape (n_samples, 2)
        Input features (2D only).
    y : np.array, shape (n_samples,)
        Binary target labels (0 or 1).
    lr : float
        Learning rate.
    epochs : int
        Number of iterations.
        
    Returns:
    m_list : list
        Slopes of decision boundary over iterations.
    b_list : list
        Intercepts of decision boundary over iterations.
    weights : np.array
        Final weights (bias, w1, w2)
    """
    # Add bias term
    X_bias = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X_bias.shape[1])
    
    m_list, b_list = [], []

    for _ in range(epochs):
        for j in range(len(X)):
            z = np.dot(X_bias[j], weights)
            y_hat = sigmoid(z)
            
            # Gradient update for logistic loss
            weights += lr * (y[j] - y_hat) * X_bias[j]
        
        # Slope and intercept for plotting decision boundary
        m_list.append(-(weights[1] / weights[2]))
        b_list.append(-(weights[0] / weights[2]))
    
    return m_list, b_list, weights
```

```python
m_list, b_list, final_weights = perceptron_sigmoid(X, y)
```
